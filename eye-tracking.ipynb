```python
# standard computer vision imports
import numpy as np
import cv2

import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# reading in files
import re, os

# displaying images inline
from IPython.display import Image, display
```

### Debugging methods

Mostly drawing functions to make outputs clearer, don't worry too much about these.

```python
# write small text with a black shadow
def pretty_txt(img, txt, (x, y), c=[255,255,255], s=0.35, t=1):
    
    (w,h), _ = cv2.getTextSize(txt, cv2.FONT_HERSHEY_SIMPLEX, s, 1)
    cv2.putText(img, str(txt), (int(x-w/2)+t,int(y+h/2)+t), cv2.FONT_HERSHEY_SIMPLEX, s, 0, t)
    cv2.putText(img, str(txt), (int(x-w/2),int(y+h/2)), cv2.FONT_HERSHEY_SIMPLEX, s, c, t)

    
# draws a polyline between points
def polyline(img, pts, c=[255,255,255], t=1, closed=False, circles=False):
    
    pts = np.array(pts).astype(int)
    if closed : pts = np.append(pts,[pts[0]],axis=0)
    for (x1,y1), (x2,y2) in zip(pts[:-1], pts[1:]):
        cv2.line(img,(x1,y1),(x2,y2), c, t)
        if circles : cv2.circle(img, (x1,y1), t*2, c, -1) 
            

# remove white margins (leave some padding)
def trim_fig(fig, pad=8):

    _, thresh = cv2.threshold(cv2.cvtColor(fig,cv2.COLOR_BGR2GRAY) , 254, 255, cv2.THRESH_BINARY_INV)
    thresh = cv2.dilate(thresh, np.ones((pad,pad),np.uint8))
    
    j, i = thresh.any(0).nonzero()[0], thresh.any(1).nonzero()[0]
    return fig[i[:,None], j[None,:]]
```

### Choose inputs

Change `img_choice` to set face image.

```python
img_choice = 6

# show face images in directory
face_imgs = [cv2.imread(f) for f in os.listdir(".") if f.startswith('img_')];
face_imgs = [cv2.resize(img, (200,200*img.shape[0]/img.shape[1])) for img in face_imgs]

# show red border around chosen face image
f_img = face_imgs[img_choice-1];
f_img[:4,:] = f_img[-4:,:] = f_img[:,:4] = f_img[:,-4:] = [0,0,255]

for i, f_img in enumerate(face_imgs):
    pretty_txt(f_img, str(i+1), (20, 20), s=1, t=2)    

# visualize the choice
cv2.imwrite("all_face_images.png", np.hstack(face_imgs))
display(Image("all_face_images.png"))
    
# set filenames for the chosen face
filename_face_img = 'img_'+str(img_choice-1)+'.png'
filename_ldmks = 'landmarks_'+str(img_choice-1)+'.txt'
filename_mean_ldmks = 'Multi-PIE_aligned_PDM_68.txt'
```

### Visualize 2D detected landmarks

```python
img_bgr = cv2.imread(filename_face_img)

# read in landmark positions from text file (ignore non-float characters)
with open(filename_ldmks) as f:
    vals = [float(re.sub("\\;|\\[|\\]", "", line)) for line in f]

# turn list of [x1,x2,...,y1,y2] into [(x1,y1),(x2,y2)...]
ldmks_2d = zip(vals[:len(vals)/2], vals[len(vals)/2:])

eye_ldmks_2d   = ldmks_2d[36:48] # both eyes together
nose_ldmks_2d  = ldmks_2d[27:36] # nose ridge and nostrils
mouth_ldmks_2d = ldmks_2d[48:]   # whole mouth

# draw landmarks on face
debug_img = np.copy(img_bgr)
for (idx,(x,y)) in enumerate(ldmks_2d):
    pretty_txt(debug_img, str(idx), (int(x),int(y)-10))
    cv2.circle(debug_img, (int(x),int(y)), 2, [255,255,0], -1)

# show the landmarks in a cropped image
l2d_xs, l2d_ys = np.array(ldmks_2d).T[0], np.array(ldmks_2d).T[1]
cv2.imwrite("2d-landmarks.png", debug_img[np.min(l2d_ys)-20:np.max(l2d_ys)+20,
                                          np.min(l2d_xs)-20:np.max(l2d_xs)+20])
display(Image("2d-landmarks.png"))
```

### Load 3D mean face model

```python
# read in mean 3D facial landmark positions
with open(filename_mean_ldmks) as f:
    vals = [float(re.sub("\\;|\\[|\\]", "", line)) for line in f]
    xs, ys, zs = vals[:len(vals)/3], vals[len(vals)/3:len(vals)/3*2], vals[len(vals)/3*2:]
    ldmks_3d = zip(xs, ys, zs)

# extract the landmarks of interest
eye0_ldmks_3d         = ldmks_3d[36:42] # left eye
eye1_ldmks_3d         = ldmks_3d[42:48] # right eye
eyebrow0_ldmks_3d     = ldmks_3d[17:22] # left eyebrow
eyebrow1_ldmks_3d     = ldmks_3d[22:27] # right eyebrow
eye_ldmks_3d          = ldmks_3d[36:48] # both eyes together
nose_ldmks_3d         = ldmks_3d[27:36] # nose ridge and nostrils
mouth_ldmks_3d        = ldmks_3d[48:]   # whole mouth
face_outline_ldmks_3d = ldmks_3d[:17]   # face outline

# set bounding points around 3D eye region [36:48] (fixed 20mm offset)
eye_bounds_mm = [(max(l) - min(l))+20 for l in [xs[36:48], ys[36:48]]]
means = [sum(l) / float(len(l)) for l in [xs[36:48], ys[36:48], zs[36:48]]]
eyes_roi_3d = [(-1,-1),(1,-1),(1,1),(-1,1)]
eyes_roi_3d = [(x*eye_bounds_mm[0]/2, y*eye_bounds_mm[1]/2) for (x,y) in eyes_roi_3d]
eyes_roi_3d = [(x+means[0], y+means[1], means[2]) for x,y in eyes_roi_3d]
```

### Visualize mean 3D face model

```python
# convenience function for OpenCV coords -> Matplotlib
def cv_to_plt(coords, closed=False):
    if closed: coords = coords+[coords[0]]
    xs, ys, zs = np.array(coords).T.tolist()
    return [xs, zs, [-y for y in ys]]

# prepare square 3D plot
fig = plt.figure(figsize=(8,8))
ax = fig.add_subplot(111, projection='3d')

# plot different face parts
ax.plot(*cv_to_plt(eye0_ldmks_3d, True)  , color='b', marker='.')
ax.plot(*cv_to_plt(eye1_ldmks_3d, True)  , color='g', marker='.')
ax.plot(*cv_to_plt(eyebrow0_ldmks_3d)    , color='r', marker='.')
ax.plot(*cv_to_plt(eyebrow1_ldmks_3d)    , color='c', marker='.')
ax.plot(*cv_to_plt(nose_ldmks_3d)        , color='m', marker='.')
ax.plot(*cv_to_plt(mouth_ldmks_3d)       , color='y', marker='.')
ax.plot(*cv_to_plt(face_outline_ldmks_3d), color='k', marker='.')
ax.plot(*cv_to_plt(eyes_roi_3d, True)    , color='y', marker='.', ls='--')

# Create cubic bounding box to force equal aspect ratio
Xb = 100*np.mgrid[-1:2:2,-1:2:2,-1:2:2][0].flatten()
Yb = 100*np.mgrid[-1:2:2,-1:2:2,-1:2:2][1].flatten()
Zb = 100*np.mgrid[-1:2:2,-1:2:2,-1:2:2][2].flatten()
for xb, yb, zb in zip(Xb, Yb, Zb):
   ax.plot([xb], [yb], [zb], 'w')

ax.view_init(elev=15, azim=-80)

# for interactive pylab: plt.draw()

# temporarily save plot for trimming
plt.savefig('3d-mean-face.png')

cv2.imwrite("3d-mean-face.png", trim_fig(cv2.imread("3d-mean-face.png"),20))
display(Image("3d-mean-face.png"))
```

```python
img_bgr = cv2.imread(filename_face_img)

# placeholder camera calibrations
# TODO: calibrate camera
cam_mat = np.matrix([[700, 0  , 800/2],
                     [0  , 700, 600/2],
                     [0  , 0  , 1    ]], dtype=np.float32)

# solve PnP to find rotation and translation of 3d face
retval, rvec, tvec = cv2.solvePnP(np.array(ldmks_3d), np.array(ldmks_2d), cam_mat, np.array([]))

# these points are the bounds of the 2d eyes roi
er2d_pts, _ = cv2.projectPoints(np.array(eyes_roi_3d, dtype=np.float32), rvec, tvec, cam_mat, np.array([]))

# draw eye features and ROI
debug_img = np.copy(img_bgr);
polyline(debug_img, np.squeeze(er2d_pts), c=[255,0,255], closed=True, circles=True)
polyline(debug_img, eye_ldmks_2d[:6], c=[255,0,255], closed=True, circles=True)
polyline(debug_img, eye_ldmks_2d[6:], c=[255,0,255], closed=True, circles=True)

cv2.imwrite("eye_roi_in_img.png", debug_img[np.min(l2d_ys)-20:np.max(l2d_ys)+20,
                                            np.min(l2d_xs)-20:np.max(l2d_xs)+20])
display(Image("eye_roi_in_img.png"))
```

```python
roi_width = 400;

# prepare a 2d region for frontalizing eyes
aspect_ratio = (eyes_roi_3d[2][1]-eyes_roi_3d[0][1]) / (eyes_roi_3d[1][0]-eyes_roi_3d[0][0])
eyes_roi_w, eyes_roi_h = roi_width, roi_width*aspect_ratio
eyes_roi_2d = [(0,0),(eyes_roi_w,0),(eyes_roi_w,eyes_roi_h),(0,eyes_roi_h)]

# warp the image eye region into the aligned ROI
# TODO: use 'getPerspectiveTransform'
h, _ = cv2.findHomography(np.squeeze(er2d_pts), np.array(eyes_roi_2d, dtype=np.float32))
eyes_roi_2d = cv2.warpPerspective(img_bgr, h, (int(eyes_roi_w), int(eyes_roi_h)))

# draw detected eye points on the frontalized eye ROI
debug_img = eyes_roi_2d.copy()
eye_ldmks_in_roi = cv2.perspectiveTransform(np.array([eye_ldmks_2d], dtype=np.float32), h)
eye_ldmks_in_roi = np.squeeze(eye_ldmks_in_roi)
for (x,y) in np.squeeze(eye_ldmks_in_roi):
    cv2.circle(debug_img, (int(x),int(y)), 2, [255,0,255], -1)

cv2.imwrite("eyes_roi_2d.png", debug_img)
display(Image("eyes_roi_2d.png"))
```
